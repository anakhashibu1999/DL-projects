```python
import cv2
import matplotlib.pyplot as plt
```

```python
img =cv2.imread('/home/Videos/opencv/face_detection/group.jpg')
```

```python
plt.imshow(img)
```

```python
converted_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(converted_img)
```

```python
face_cascade=cv2.CascadeClassifier('/home/Videos/computer_vision/opencv/face_detection/haarcascade_frontalface_default.xml')
```

```python
face_cascade
```

```python
img_grey=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
img_grey
```

```python
face = face_cascade.detectMultiScale(img_grey,scaleFactor=1.1,minNeighbors=4)
```

```python
face
```

```python
len(face)
```

```python
for ( x,y,w,h) in face:
    cv2.rectangle(converted_img,(x,y),(x+w,y+h),color=(0,255,0),thickness=3)

```

```python
plt.imshow(converted_img)
```

```python

video=cv2.VideoCapture('/home/Videos/computer_vision/opencv/face_detection/mh.mp4')
while True:
    sucess,frame =video.read()

    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)

    faces=face_cascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=4)

    print(len(faces))

    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),3)


    cv2.imshow('webcam',frame)
    if cv2.waitKey(1) & 0xFF ==27:
        break
video.release()
cv2.destroyAllWindows()
```

```python
video=cv2.VideoCapture('/home/Videos/computer_vision/opencv/face_detection/mh.mp4')
while True:
    success,frame =video.read()

    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)

    faces=face_cascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=4)
    eyes=eye_cascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=4)
    # print(len(faces))
    # print(len(eyes))

    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),3)

    for (x,y,w,h) in eyes:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),3)
        

    cv2.imshow('video',frame)
    if cv2.waitKey(1) & 0xFF==27:
        break
video.release()
cv2.destroyAllWindows()



```
